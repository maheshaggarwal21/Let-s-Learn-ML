# Polynomial Regression

- general equation -> y = a0 + a1* x^1 + a2* x^2 + a3* x^3 + ... + an* x^n

## Data Preprocessing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("Position_Salaries.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
X = data.iloc[:,1:-1].values
y = data.iloc[:,-1].values
```

## Training The Linear Regression Model On Whole data

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

## Training The Polynomial Regression Model On Whole Data

```python
from sklearn.preprocessing import PolynomialFeatures
poly_mod = PolynomialFeatures(degree=4)
X_poly = poly_mod.fit_transform(X)
lin_reg_poly = LinearRegression()
lin_reg_poly.fit(X_poly,y)
```

## Visualising Linear Regression Results

```python
plt.scatter(X,y,color = 'red')
plt.plot(X, lin_reg.predict(X),color='blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

## Visualising Polynomial Regression Results

```python
plt.scatter(X,y,color = 'red')
plt.plot(X, lin_reg_poly.predict(X_poly),color='blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

## Predicting A New Result Using Both Models

```python
y_poly = lin_reg_poly.predict(poly_mod.fit_transform([[6.5]]))
y_lin = lin_reg.predict([[6.5]])
```

```python
print(y_poly,y_lin)
```
